## Step 0 — Import Libraries

```python
import pandas as pd
import numpy as np

print("Pandas version :", pd.__version__)
print("NumPy  version :", np.__version__)
```

## Step 1 — Load the CSV Dataset into a DataFrame
We use `pd.read_csv()` to load the raw shopping data file.


```python
df = pd.read_csv('shopping_data.csv')

print(f"✅ Dataset loaded — {df.shape[0]} rows × {df.shape[1]} columns")
df
```

## Step 2 — Explore the Data

```python
# First 5 rows
print("=== HEAD (first 5 rows) ===")
df.head()
```

```python
# Last 5 rows
print("=== TAIL (last 5 rows) ===")
df.tail()
```

```python
# Shape, columns, dtypes
print("=== SHAPE ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n=== COLUMNS ===")
print(df.columns.tolist())

print("\n=== DATA TYPES ===")
print(df.dtypes)
```

```python
# Statistical summary
print("=== DESCRIBE ===")
df.describe()
```

```python
# Full info (non-null counts)
print("=== INFO ===")
df.info()
```

## Step 3 — Handle Missing Values
Identify nulls, then fill or drop them.


```python
# Identify missing values
print("=== MISSING VALUES ===")
missing = df.isnull().sum()
pct = (missing / len(df) * 100).round(1)
print(pd.DataFrame({'Count': missing, '%': pct})[missing > 0])
print(f"\nTotal missing cells: {df.isnull().sum().sum()}")
```

```python
# Fill missing price with median
price_median = df['price'].median()
df['price'] = df['price'].fillna(price_median)
print(f"✅ Filled missing 'price' with median → {price_median:.2f}")

# Fill missing quantity with mode (most common value)
qty_mode = df['quantity'].mode()[0]
df['quantity'] = df['quantity'].fillna(qty_mode)
print(f"✅ Filled missing 'quantity' with mode → {qty_mode}")

# Fill missing size with 'Unknown'
df['size'] = df['size'].fillna('Unknown')
print("✅ Filled missing 'size' with 'Unknown'")
```

```python
# Verify — should be 0
print(f"Remaining missing values: {df.isnull().sum().sum()}")
df.head()
```

## Step 4 — Filter Rows & Select Columns
Use boolean indexing to extract meaningful subsets.


```python
# Filter: rows where price > 500
print("=== Orders with price > 500 ===")
df[df['price'] > 500][['order_id', 'product', 'category', 'price']]
```

```python
# Filter: specific categories using .isin()
print("=== Electronics & Beauty orders ===")
df[df['category'].isin(['Electronics', 'Beauty'])][['order_id', 'product', 'category', 'price', 'rating']]
```

```python
# Filter: high-rated products (rating >= 4.0)
print("=== High-rated products (rating >= 4.0) ===")
df[df['rating'] >= 4.0][['order_id', 'product', 'category', 'rating']]
```

```python
# Select specific columns only
print("=== Selected columns: order_id, product, price, quantity ===")
df[['order_id', 'product', 'price', 'quantity']]
```

## Step 5 — Remove Duplicates


```python
# Detect duplicates
print(f"Duplicate rows found: {df.duplicated().sum()}")
df[df.duplicated()]
```

```python
# Drop duplicates
rows_before = len(df)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"✅ Removed {rows_before - len(df)} duplicate(s)")
print(f"   Rows before : {rows_before}")
print(f"   Rows after  : {len(df)}")
```

## Step 6 — Create Derived Column: `total_amount`
**Formula:** `total_amount = price × quantity`


```python
# Ensure correct dtypes first
df['price']    = pd.to_numeric(df['price'],    errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce').fillna(df['quantity'].mode()[0]).astype(int)

# Create derived column
df['total_amount'] = (df['price'] * df['quantity']).round(2)

print("=== Sample with total_amount ===")
df[['order_id', 'product', 'price', 'quantity', 'total_amount']]
```

```python
# Quick stats on total_amount
print("=== total_amount Statistics ===")
print(df['total_amount'].describe().round(2))
print(f"\nTop revenue category:")
print(df.groupby('category')['total_amount'].sum().sort_values(ascending=False).round(2))
```

## Step 7 — Save the Cleaned Dataset


```python
output_path = 'shopping_cleaned.csv'
df.to_csv(output_path, index=False)
print(f"✅ Cleaned dataset saved → '{output_path}'")
print(f"   Final shape : {df.shape[0]} rows × {df.shape[1]} columns")
```
